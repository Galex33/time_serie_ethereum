```python
from f_global import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ethereum_day = read_copy_csv("data/ETHUSD_day.csv")
```

```python
def df_info(df):
        len_df = len(df)
        all_columns = len(df.columns)

        print(f"""
        Longueur du dataset : {len_df} enregistrements
        Nombre de colonnes : {all_columns}
        """)

        echantillonColonnes = []
        for i in df.columns:
            listcolumn = str(list(df[i].head(5)))
            echantillonColonnes.append(listcolumn)
       
        pd.set_option("max_rows", None)
        obs = pd.DataFrame({'type': list(df.dtypes),
        'Echantillon': echantillonColonnes,
        "% de valeurs nulles":
        round(df.isna().sum() / len_df * 100, 2),
        'Nbr L dupliquées' : (df.duplicated()).sum(),
        'Nbr V unique' : df.nunique()
        })
        return obs
```

```python
ethereum_day = ethereum_day.drop(columns=['Unix Timestamp', 'Symbol'])
ethereum_day.columns = ['Date', 'Opening_Price_ETH', 'Highest_rice_ETH', 'Lowest_Price_ETH', 'Lowest_Price_ETH', 'Vol_ETH']
```

```python
ethereum_day['Date'] =ethereum_day['Date'].apply(lambda x:x.replace(' 04:00', ''))
ethereum_day['Date'] = pd.to_datetime(ethereum_day['Date'], dayfirst=True )
ethereum_day = ethereum_day.set_index('Date')
ethereum_day = ethereum_day.sort_index(axis = 0, ascending = True)
ethereum_day
```

```python
df_info(ethereum_day)
```

```python
ethereum_day_opening_price = ethereum_day[['Opening_Price_ETH']]
ethereum_day_opening_price
```

```python
ethereum_day_opening_price
```

```python
plt.figure(figsize=(12, 8))
ethereum_day_opening_price.loc['2016' : '2021', 'Opening_Price_ETH'].plot()
ethereum_day_opening_price.loc['2016' : '2021', 'Opening_Price_ETH'].resample('W').mean().plot(label='moyenne par semaine', lw=2, ls='--', alpha=0.8)
ethereum_day_opening_price.loc['2016' : '2021', 'Opening_Price_ETH'].resample('M').mean().plot(label='moyenne par mois', lw=3, ls=':', alpha=0.8)
plt.legend()
plt.ylabel('USD$', fontsize=18)
plt.xlabel('YEAR-MONTH', fontsize=18)
plt.show()
```
